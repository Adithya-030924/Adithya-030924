```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
d2['airport']=le.fit_transform(d2['airport'])
d2['waterbody']=le.fit_transform(d2['waterbody'])
d2['bus_ter']=le.fit_transform(d2['bus_ter'])
```

```python
x2=d2.drop(['price'],axis=1)
y2=d2['price']
```

```python
from sklearn.model_selection import train_test_split
x2_train,x2_test,y2_train,y2_test=train_test_split(x2,y2,test_size=0.2,random_state=100)
```

```python
from sklearn.preprocessing import StandardScaler
sd=StandardScaler()
x2_train1=sd.fit_transform(x2_train)
x2_test1=sd.transform(x2_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x2_train1,y2_train)
```

```python
y_pred2=lr.predict(x2_test1)
```

```python
from sklearn.metrics import accuracy_score, mean_squared_error
import numpy as np
acc2=lr.score(x2_test1,y2_test)
mse2=mean_squared_error(y2_test,y_pred2)
acc2,np.sqrt(mse2),mse2
```
