```python
from scipy import stats
numerical_df = df.select_dtypes(include=np.number)
z_scores = np.abs(stats.zscore(numerical_df))
threshold = 3
df_no_outliers = df[(z_scores < threshold).all(axis=1)]
```

```python
df_no_outliers
```

```python
plt.figure(figsize=(25,12))
plt.boxplot(df_no_outliers)
plt.ylabel('Values')
plt.title('Boxplot of Features')
plt.xticks(ticks=range(1, len(df_no_outliers.columns) + 1), labels=numerical_df.columns, rotation=90)
plt.show()
```

```python
x=df_no_outliers.drop('SOURCE', axis=1)
y=df_no_outliers['SOURCE']
```

### **without** **pipeline**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
```

```python
print('accuracy :' ,accuracy)
print('confusion matrix :',confusion_matrix(y_test,y_pred))
print('classification report :\n',classification_report(y_test,y_pred))
```

```python
feature_imp=pd.Series(model.feature_importances_,index=df_no_outliers.drop('SOURCE', axis=1).columns).sort_values(ascending=False)
feature_imp.sort_values(ascending=False).plot(kind='bar',title='Feature Importance')
plt.show()
```

```python
num_features = df.select_dtypes(include=np.number).columns.tolist()
num_features.remove('SOURCE')
print(num_features)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV


num_pipeline = Pipeline([

    ("scaler", StandardScaler())
])


preprocessor = ColumnTransformer([
    ("num", num_pipeline, num_features)
])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
pipeline = Pipeline([
    ("preprocessing", preprocessor),
    ("classifier", LogisticRegression())
])
param_grid = [
    {
        "classifier": [LogisticRegression(max_iter=1000)],
        "classifier__C": [0.1, 1, 10]
    },
    {
        "classifier": [RandomForestClassifier(random_state=42)],
        "classifier__n_estimators": [100, 200],
        "classifier__max_depth": [None, 10, 20]
    },
    {
        "classifier": [SVC()],
        "classifier__kernel": ['linear', 'rbf'],
        "classifier__C": [1, 10]
    }
]

# Grid Search with 5-fold cross-validation
grid = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)
grid.fit(x_train, y_train)
```

```python
print("Best model:", grid.best_estimator_)
print("Best parameters:", grid.best_params_)

y_pred = grid.predict(x_test)

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
